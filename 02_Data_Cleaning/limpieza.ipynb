```python
import pandas as pd

# # Si 'limpieza.ipynb' y 'datos_sucios.csv' están en la misma carpeta:
df = pd.read_csv('datos_sucios.csv')

print("--- Datos originales con errores ---")
print(df)

# Aquí veremos qué datos faltan
print("\n--- ¿Cuántos datos nulos hay por columna? ---")
print(df.isnull().sum())

# 1. Rellenar los huecos de Stock con 0 (si no hay datos, asumimos que no hay stock)
df['Stock'] = df['Stock'].fillna(0)

# 2. Rellenar los huecos de Precio con el promedio de los precios existentes
precio_promedio = df['Precio'].mean()
df['Precio'] = df['Precio'].fillna(precio_promedio)

print("\n--- Dataset después de la limpieza ---")
print(df)
```

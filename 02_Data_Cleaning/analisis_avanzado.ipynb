```python
# 1. Limpiar el simbolo '$' de la columna Precio y convbertirla a numero
# .str.replace('$'.'') quita e simbolo y .astype(float) convierte a numero
df['Precio'] = df['Precio'].replace('[\$,]', '', regex=True).astype(float)

# Usamos 'mixed' para que Pandas infiera el formato de cada fecha indivualmente
# 2. Convertir la columna 'Fecha de Publicación' a formato datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], format='mixed')

# 3. Ver como quedaron los tipos de datos ahora
print(df.info())
print("\n--- Vista previa de los datos limpios ---")
print(df.head())


```

```python
# 1. Agrupar por 'Producto' y sumar el stock
stock_por_producto = df.groupby('Producto')['Stock'].sum()
print("--- Stock Total por Producto ---")
print(stock_por_producto)

# 2. Agrupar por 'Producto' y calcular el precio promedio
precio_promedio = df.groupby('Producto')['Precio'].mean()
print("\n--- Precio Promedio por Producto ---")
print(precio_promedio)
```

```python
import matplotlib.pyplot as plt

# Usamos la serie de stock que calculamos en el Nivel 2
stock_por_producto.plot(kind='bar', color='skyblue', edgecolor='black', title='Stock Total por Producto')

# Mejoramos un poco el gráfico
plt.ylabel('Cantidad en Stock')
plt.xlabel('Producto')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añade líneas de guía horizontales

# Mostramos el gráfico
plt.show()
```

```python
# 1. Veamos cuántos nulos tenemos por columna
print("--- Valores nulos antes de limpiar ---")
print(df.isnull().sum())

# 2. Imputación inteligente: rellenar los precios vacíos con el promedio del producto
# Esto evita que perdamos filas enteras por un dato faltante
df['Precio'] = df.groupby('Producto')['Precio'].transform(lambda x: x.fillna(x.mean()))

# 3. Rellenar los valores nulos de Stock con 0 (si no hay, asumimos 0)
df['Stock'] = df['Stock'].fillna(0)

print("\n--- Valores nulos después de limpiar ---")
print(df.isnull().sum())
```

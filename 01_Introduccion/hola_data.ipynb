```python
import pandas as pd

# 1. Creamos un pequeño dataset (como si fuera una tabla de excel)
data = {
    'Producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares'],
    'Precio': [1000, 50, 80, 300, 150],
    'Cantidad': [5, 20, 15, 10, 8]
}

df=pd.DataFrame(data)

# 2. Mostramos el dataset (df.head muestra las primeras filas)
print("--- Primeras filas del dataset ---")
print(df.head())

# 3. Filtracion de datos (ejemplo: productos con precio mayor a 100)
caros = df[df['Precio'] > 100]
print("\n--- Productos con precio mayor a 100 ---")
print(caros)
# 4. Agregar una nueva columna (ejemplo: Total = Precio * Cantidad)
df['Total'] = df['Precio'] * df['Cantidad']
print("\n--- Dataset con columna Total ---")
print(df)

# 5. Exportar el resultado a un archivo CSV
df.to_csv('resultados_ventas.csv', index=False)
print("\n¡Archivo 'resultados_ventas.csv' creado con éxito!")
```
